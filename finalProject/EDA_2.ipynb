```python
#finally, peace PEACE!
import pandas as pd

import plotly.express as px

import seaborn as sb
```

```python
data = pd.read_csv("./Data/water_potability.csv")
data.head()
```

```python
fig = px.parallel_categories(data, color="Conductivity", color_continuous_scale=px.colors.sequential.Inferno)
fig.show()
```

```python
fig = px.ecdf(data, x="ph", color="Potability")
fig.show()
```
